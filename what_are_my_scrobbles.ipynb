```
import pandas as pd
import glob, os    
import requests
import json
from time import sleep
from random import randint
from pandas.io.json import json_normalize
```

```
def buildit(uri):
    """
    builds a list urls and filenames
    feed this to grabit which will 
    parse the results and generate files
    """
    r=requests.get(uri)
    if r.status_code == 200:
        print('OK') 
        data = json.loads(r.text)
        totPages=int(data['recenttracks']['@attr']['totalPages'])
        pages=[]
        for i in range(1,totPages):
            page=[]
            page.append(str(i)+'.csv')
            page.append(uri+'&page='+str(i))
            pages.append(page)
    return pages

def grabit(outerList):
    """
    takes a list of urls/filenames
    generates files 
    """
    for innerList in outerList:
        url=innerList[1]
        fn= innerList[0] 
        j=randint(1,9) # ratelimit appears to be 500 per 5 minutes 
        sleep(j)       # this keeps us comfortably below that 
        r=requests.get(url)
        while r.status_code != 200:
          j=randint(1,9)
          sleep(j)
          r=requests.get(url)
        data = json.loads(r.text)
        ddf=json_normalize(data['recenttracks']['track'])
        ddf.to_csv(fn, index=False)
        print(fn+" saved")

    print("all Done!")
```

```
#alter the below to match your details
user='yourLastFm username'
api_key='get your api from https://secure.last.fm/login?next=/api/account/create' 
file_name='the name you want your final dataset to be saved as.csv' 
```

```
#build the list, download the files
url='http://ws.audioscrobbler.com/2.0/?method=user.getrecenttracks&user='+user+'&api_key='+api_key+'&format=json'

urlList=buildit(url)

grabit(urlList)
```

```
# this will grab your scrobbles 50 at a time.  
# it may take a while to grab them all, it took me ~4 hours to get 180k
# additionally, you may need to restart the process; it seems to stall
# every ~100 minutes
```

```
# assuming we have successfully grabbed and saved all the scrobbles:
#lets munge them into a single dataframe:

df = pd.concat(map(pd.read_csv, glob.glob(os.path.join('', "*.csv"))))

```

```
# wholly possible that we have downloaded 
# the same scrobbles multiple times if we have 
# had to start/stop the download based on timeouts.
# let's get rid of explicit dupes:

ddf=df.drop_duplicates()
```

```
ddf.to_csv(file_name, index=False)
```
